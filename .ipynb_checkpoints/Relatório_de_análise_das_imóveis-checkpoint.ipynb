## 5.2 analisando o ano de construção dos imóveis com o preço de compra

```python
dados
```

```python
# observando os imóveis atrvés do seu ano de construção dos imóveis
ano_de_construcao = dados.groupby(['AnoConstrucao'])['Preco'].mean().round(2)
ano_de_construcao_preco = ano_de_construcao.to_frame()
ano_de_construcao_preco.rename(columns={'Preco': 'Preço'}, inplace = True)
ano_de_construcao_preco
```

```python
# Gráfrico de barras e de linha com a média dos valores dos imóveis por ano de construção

# definindo o tamanho das figuras
fig = plt.figure(figsize=(35, 28))

# gráfico de barras
fig.add_subplot(3, 1, 1)
ax = sns.barplot(data = ano_de_construcao_preco, y='Preço', x=ano_de_construcao_preco.index, orient='v', palette = 'Blues_d')
ax.figure.set_size_inches(12,8)
ax.set_title('Média do preço de venda dos imóveis por ano de contrução', fontsize=15, loc='left')
ax.set_xlabel('Ano de construção', fontsize=11)
ax.set_ylabel('Preço (dólar)', fontsize=11)
plt.xticks(rotation=75);

# gráfico de linha
fig.add_subplot(3, 1, 3)
ax = sns.lineplot(data = ano_de_construcao_preco, y='Preço', x=ano_de_construcao_preco.index, palette = 'Blues_d')
ax.figure.set_size_inches(12,9)
ax.set_title('Média do preço de venda dos imóveis por ano de contrução', fontsize=15, loc='left')
ax.set_xlabel('Ano de construção', fontsize=11)
ax.set_ylabel('Preço (dólar)', fontsize=11)
plt.xticks(rotation=45);
```

```python
# criando nova variável no conjunto de dados (1-> depois de 1990 e 0->antes de 1990)
dados['depois1990'] = dados['AnoConstrucao'].apply(lambda x: 1 if x >= 1990 else 0)
dados['depois1990'].replace({0: 'Antes de 1990', 1: 'Depois de 1990'}, inplace=True)
dados.head(2)
```

```python
# observando distribuição do preço de vendas de imóveis construídos antes e depois de 1990
boxplot = sns.boxplot(data=dados, x='depois1990', y='Preco', width=0.8, palette = 'mako')
boxplot.figure.set_size_inches(10,5)
boxplot.set_title('Boxplot do preço de imóveis construídos antes e depois de 1990', loc = 'left', fontsize=15)
boxplot.set_ylabel('Preço (dólar)', fontsize=14)
boxplot.set_xlabel('Construção', fontsize=14);
```

```python
# Construindo tabela de estatísticas dos preços dos imóveis construídos antes e depois de 1990 
columns = {
    'count': 'Frequência',
    'mean': 'média',
    'std': 'desvio padrão'
}
tabela_estatisticas_construcao = pd.DataFrame(data = dados.groupby(['depois1990'])['Preco'].describe().round(2))
tabela_estatisticas_construcao.rename(columns = columns, inplace = True)
tabela_estatisticas_construcao.index.name = 'Estatísticas'
tabela_estatisticas_construcao.T
```

<font style="color: green" size=3>A partir dos gráficos e das informações extraídas acima, é obseravdo que imóveis que foram construídos mais recentemente tendem a apresentar um valor de preço de compra maior. Através do boxplot e da tabela percebemos que 75% dos imóveis que foram construídos depois de 1990 apresentam valor de compra maior que 159.671,25 dólares e os que foram construídos antes de 1990 apresentam um valor de compra maior que 123.000,00 doláres. Além disso, é visível que a média dos preços de compra dos imóveis que foram construídos depois de 1990 é maior que a média dos imóveis construídos antes de 1990.</font>

<font style="color: blue" size=3> Com toda análise realizada acima, irei aceitar a nossa hipótese Nº1 de que as casas que foram contruídas mais recentemente tendem a apresentar um preço de compra mais elevado.</font>

## 6.2 Gráficos de dispersão da variável dependente (y) com as variáveis explicatiavas (x)

```python
# realizando um função que plota um gráfico de dispersão
def dispersao(df, x, y, xlabel, ylabel, titulo, ajuste_x = 0.45, ajuste_y = 1.02):
    ax = sns.lmplot(data = df, x = x, y = y)
    ax.fig.set_size_inches(5,5)
    plt.suptitle(titulo, fontsize = 14, x = ajuste_x, y = ajuste_y)
    plt.xlabel(xlabel, fontsize = 13)
    plt.ylabel(ylabel, fontsize = 13);
```

##### <font style="font-size: 20px;">Teste de significância individual das features</font>

## 8.2 Estimando o modelo baseline (DummyRegressor) baseado na média de preços de compra

```python
from sklearn.dummy import DummyRegressor
```

```python
# criando o modelo baseline
modelo_baseline = DummyRegressor(strategy = 'mean')

# treinando o modelo baseline
modelo_baseline.fit(x_train, y_train)

# obtendo as previsões do modelo baseline
y_previsto_baseline = modelo_baseline.predict(x_test)

# observando as métricas
r2_baseline = metrics.r2_score(y_test, y_previsto_baseline)
mae_baseline = metrics.mean_absolute_error(y_test, y_previsto_baseline)
mape_baseline = np.mean( np.abs(y_test - y_previsto_baseline) / y_test ) * 100
rmse_baseline = np.sqrt(metrics.mean_squared_error(y_test, y_previsto_baseline))

# tabela de métricas
pd.DataFrame([np.abs(r2_baseline), mae_baseline, mape_baseline, rmse_baseline]).rename(index = {0: 'R2 Baseline',1: 'MAE Baseline', 2: 'MAPE Baseline (%)', 3: 'RMSE Baseline'},columns = {0: 'Métricas'}).round(2)
```

<font style="color: green" size=3>Com a tabela de métricas acima é observado que a forma como a empresa vem precificando os imóveis, através de uma simples média do valor do mercado e intuição de mercado (chute), não explica muito bem a variabilidade dos dados. Além disso é visível que o erro absoluto percentual médio é de 17,4% que é um valor bem alto, ou seja a forma atual que a empresa esta precificando os imóveis erra em média 17,4% o valor do imóvel, o que pode causar bastante prejuízo ao faturamento da empresa.</font>

## 8.3 estimando o modelo de regressão linear 

```python
# como já temos os dados divididos em treino e teste agora é só criar e treinar o modelo de regressão linear

# criando o modelo
modelo = LinearRegression()

# treinando o modelo
modelo.fit(x_train, y_train)

# previsões do modelo
y_previsto = modelo.predict(x_test)

# Observando métricas
r2 = metrics.r2_score(y_test, y_previsto)
mae = metrics.mean_absolute_error(y_test, y_previsto)
mape = np.mean( np.abs(y_test - y_previsto) / y_test ) * 100
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_previsto))

# tabela de métricas
pd.DataFrame([r2, mae, mape, rmse]).rename(index = {0: 'R2', 1: 'MAE', 2: 'MAPE (%)', 3: 'RMSE'}, columns = {0: 'Métricas'}).round(2)
```

```python
# tabela de coeficientes do modelo de regressão
df = pd.DataFrame(modelo.coef_.round(2)).rename(columns = {0: 'Coeficientes do modelo'}, index = {
    0: 'Area',
    1: 'Nível de Qualidade do acabamento',
    2: 'Ano de Construção',
    3: 'N° de banheiros',
    4: 'Cômodos',
    5: 'Lareiras',
    6: 'Garagem',
    7: 'Zona residêncial'
})
df.loc[8] = modelo.intercept_
df.rename(index = {8: 'intercepto'}).round(2)
```

<font style="color: green" size=3>Agora com o novo modelo de precificar os imóveis, criado por mim cientista de dados da empresa, é observado um ganho de performance bem elevado na precificação dos imóveis. Com a tabela de métricas acima fica evidente que com o novo modelo temos apenas um erro absoluto percentual médio (MAPE) de 8,55%, ou seja o modelo erra em média 8,55% do valor do imóvel em sua predição, que é um valor bem menor do que a forma como a empresa vem precificando atualmente. Com o novo modelo a empresa pode aumentar o faturamento, tendo em vista que modelo chega bem perto do valor real do imóvel, gastando muito menos na compra de um imóvel do que anteriormente.</font>

<font style="color: blue" size=3>Portanto, é observado que o erro médio absoluto do novo modelo (regressão Linear) é 49,04% menor que o erro médio absoluto do modelo anterior, que é a forma como a empresa vem precificando os imóveis atualmente.</font>
